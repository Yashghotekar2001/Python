```python
###B. Find the most-experienced employee(s) per department
```

```python
# 1) per-department maximum years
dept_max = df.groupby('dept', as_index=False)['timeSpent.company'].max().rename(columns={'timeSpent.company':'max_years'})
dept_max = dept_max.sort_values('max_years', ascending=False)
print(dept_max)

# 2) merge to get full employee rows who match the per-dept max (will return ties)
dept_max_employees = dept_max.merge(df, left_on=['dept','max_years'], right_on=['dept','timeSpent.company'], how='left')

# show columns of interest
dept_max_employees[['dept','max_years','timeSpent.company','salary','satisfactoryLevel','lastEvaluation','numberOfProjects','avgMonthlyHours','left']]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.hist(df['timeSpent.company'], bins=range(df['timeSpent.company'].min(), df['timeSpent.company'].max()+2))
plt.xlabel('Years at company')
plt.ylabel('Count')
plt.title('Distribution of years at company')
plt.show()
```

```python
exp_vs_left = df.groupby('timeSpent.company', as_index=False)['left'].agg(['mean','count']).reset_index()
exp_vs_left.rename(columns={'mean':'attrition_rate','count':'n'}, inplace=True)

# plot attrition rate (but check n before trusting points with very small n)
plt.figure(figsize=(8,4))
plt.plot(exp_vs_left['timeSpent.company'], exp_vs_left['attrition_rate'], marker='o')
plt.xlabel('Years at company')
plt.ylabel('Attrition rate (fraction who left)')
plt.title('Attrition rate vs years at company')
plt.grid(True)
plt.show()
```

```python
y = df['left']
```

```python
X = df[['satisfactoryLevel','lastEvaluation','numberOfProjects',
        'avgMonthlyHours','timeSpent.company','workAccident',
        'promotionInLast5years',]]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, random_state=42, stratify=y
)

print("Training set:", X_train.shape, y_train.shape)
print("Testing set:", X_test.shape, y_test.shape)
```

```python
print("Train set distribution:\n", y_train.value_counts(normalize=True))
print("Test set distribution:\n", y_test.value_counts(normalize=True))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Initialize and train model
model = LogisticRegression(max_iter=1000)
model.fit(X_train, y_train)
```

```python
# Predictions on test set
y_pred = model.predict(X_test)
```

```python
# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Precision, Recall, F1-score
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Stayed','Left'], yticklabels=['Stayed','Left'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()
```
