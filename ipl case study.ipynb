```python
batting_first=ipl[ipl["Win_Type"]!="wickets"]
```

```python
batting_first.head()
```

```python
plt.figure(figsize=(5,5))
plt.hist(batting_first["Win_Margin"])
plt.title("Distribution of Runs")
plt.xlabel("runs")
plt.show()
```

```python
batting_first["match_winner"].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(batting_first["match_winner"].value_counts()[0:3].keys()),list(batting_first["match_winner"].value_counts()[0:3]),color=["blue","orange","red"])
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.pie(list(batting_first["match_winner"].value_counts()),labels=list(batting_first["match_winner"].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
