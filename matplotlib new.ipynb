```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
iris=sns.load_dataset("iris")
```

```python
sns.histplot(x='petal_width',data=iris)
plt.show()
```

```python
sns.histplot(x='petal_length',data=iris)
plt.show()
```

```python
iris.head()
```

```python
sns.histplot(x='petal_width',data=iris,hue='species',kde=True)
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
sns.distplot(iris['petal_width'],color='r')
plt.show()
```

```python
sns.pairplot(data=iris,hue='species')
plt.show()
```

```python
sns.catplot(x='sepal_length',data=iris,hue='species',kind='violin')
plt.show()
```

```python
sns.catplot(x='species',y='sepal_length',data=iris,kind='swarm')
plt.show()
```

```python
tc=iris.corr(numeric_only=True)
tc
```

```python
sns.heatmap(tc,annot=True)
plt.show()
```
